```python
movies_clean = movies.copy()
tv_shows_clean = tv_shows.copy()
```

```python
columns_to_drop = ["Unnamed: 0", "ID", "IMDb", "Type"]

movies_clean.drop(columns=[c for c in columns_to_drop if c in movies_clean.columns], inplace=True)
tv_shows_clean.drop(columns=[c for c in columns_to_drop if c in tv_shows_clean], inplace=True)
```

```python
def clean_rotten_tomatoes(score):
    if pd.isna(score):
        return None
    return int(score.split("/")[0])
```

```python
movies_clean["Rotten_Tomatoes"] = movies_clean["Rotten Tomatoes"].apply(clean_rotten_tomatoes)
tv_shows_clean["Rotten_Tomatoes"] = tv_shows_clean["Rotten Tomatoes"].apply(clean_rotten_tomatoes)
```

```python
movies_clean.drop(columns=["Rotten Tomatoes"], inplace=True, errors="ignore")
tv_shows_clean.drop(columns=["Rotten Tomatoes"], inplace=True, errors="ignore")
```

```python
platform_columns = ["Netflix", "Hulu", "Prime Video", "Disney+"]
```

```python
movies_unpivot = movies_clean.melt(
    id_vars= [c for c in movies_clean.columns if c not in platform_columns],
    value_vars= platform_columns,
    var_name= "Platform",
    value_name = "is_available"
)
```

```python
movies_final = movies_unpivot[movies_unpivot["is_available"] == 1]
```

```python
movies_final = movies_final.drop(columns=['is_available'])
```

```python
def search_tmdb(title, content_type, year):
    if pd.isna(title) or pd.isna(content_type):
        return None

    ct_mapped = content_type.lower().strip()
    if ct_mapped not in ["movie", "tv"]:
        return NotImplemented

    url = f"https://api.themoviedb.org/3/search/{ct_mapped}"

    params = {
        "api_key": TMDB_API_KEY,
        "query": title,
        "year": int(year) if not pd.isna(year) else None
    }

    try:
        r = requests.get(url, params=params, timeout=5)
        if r.status_code != 200:
            return None

        results = r.json().get("results", [])
        if not results:
            return None

        best = max(results, key=lambda x: x.get("vote_count", 0))

        return{
            "tmdb_id": best.get("id"),
            "tmdb_rating": best.get("vote_average"),
            "tmdb_vote_count": best.get("vote_count"),
            "tmdb_popularity": best.get("popularity"),
            "tmdb_genres": best.get("genre_ids")
        }

    except Exception:
        return None 
```

```python
import time

for batch_num in range(START_BATCH, num_batches):
    tmdb_results = []
    start_idx = batch_num * BATCH_SIZE
    end_idx = min(start_idx + BATCH_SIZE, total_rows)

    print(f"Processing batch {batch_num + 1}/{num_batches} "
          f"(rows {start_idx} to {end_idx - 1})")

    batch_df = content_master.iloc[start_idx:end_idx]

    for _, row in batch_df.iterrows():
        result = search_tmdb(
            title=row["Title"],
            content_type=row["Content_type"],
            year=row["Year"]
        )

        if result:
            result["row_id"] = row["row_id"]
        else:
            result = {
                "row_id": row["row_id"],
                "tmdb_id": None,
                "tmdb_rating": None,
                "tmdb_vote_count": None,
                "tmdb_popularity": None,
                "tmdb_genres": None
            }

        tmdb_results.append(result)

        time.sleep(0.3)


    batch_results_df = pd.DataFrame(tmdb_results)
    batch_results_df.to_csv(
        "tmdb_enrichment_progress.csv",
        mode="a",
        header=not os.path.exists("tmdb_enrichment_progress.csv"),
        index=False
    )

    print(f"Batch {batch_num + 1} saved successfully.\n")
```
